```python
import json
import os
import pandas as pd


def make_csv(base_dir, f_name):
    dir_list = os.listdir(base_dir)
    dir_list.sort()

    df = pd.DataFrame(columns=['gloss', 'spoken'])

    for i in dir_list:
        if i == '.DS_Store': continue 
        for j in os.listdir(os.path.join(base_dir,i)):
            if j == '.DS_Store': continue 
            for k in os.listdir(os.path.join(base_dir,i,j)):
                if k == '.DS_Store': continue 
                for l in os.listdir(os.path.join(base_dir,i,j,k)):
                    with open(os.path.join(base_dir,i,j,k,l)) as f: json_object = json.load(f)
                    gloss = []
                    try:
                        for nn in ['sign_gestures_both', 'sign_gestures_strong', 'sign_gestures_weak']:
                            for m in json_object['sign_script'][nn]:
                                mm = m['gloss_id'].strip()
                                if ':' in mm: qqq = mm.split(':')[-1]
                                elif mm.isdigit(): qqq = mm
                                elif mm[-1].isdigit(): qqq = mm[:-1]
                                elif mm[-1] == '#': qqq = mm[:-2]
                                else: qqq = mm
                                
                                if qqq in '하이픈': qqq = '-'
                                if qqq == '마침표': qqq = '.'
                                if qqq == '물결표': qqq = '~'
                                gloss.append((m['start'], qqq))
                    except:
                        continue
                    gloss.sort(key=lambda x: x[0])
                    new_gloss = [i[1] for i in gloss]
                    new_gloss.append('.')

                    for rr in range(len(new_gloss)-1):
                        if new_gloss[rr] == new_gloss[rr+1]: new_gloss[rr] = ''
                    
                    df.loc[len(df)] = [' '.join(new_gloss), json_object['korean_text']]
                print(f'{k} is done')
            print(f'----{j} is done')
        print(f'--------{i} is done')
    df.to_csv(f_name, index=False)

base_dir = 'path'
make_csv(os.path.join(base_dir, '1.tact_morpheme'), 'NIASL_TRN_TACT.csv')
make_csv(os.path.join(base_dir, '2.untact_morpheme'), 'NIASL_TRN_UNTACT.csv')
```
