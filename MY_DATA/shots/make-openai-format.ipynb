```python
import pandas as pd
import json

# df = pd.read_csv('../gloss_from_book.csv')
# df = df.sample(frac=1).reset_index(drop=True)

df = pd.read_csv('gloss2text/0-shot.csv')
```

```python
import random

row_data = []
for i in range(len(df)):
    line = df.iloc[i]
    # row_data.append({"prompt":f"한국 수어 시퀸스를 한국말로 번역해줘\n"
    #                           f"입력: {line['gloss']} \n ->",
    #                  "completion": f"{line['spoken']}"})
    row_data.append(
        {
            "messages": [
                {"role": "system", "content": "너는 한국 수어 번역기야. 예시처럼 말을 자연스럽게 이으면 돼\n"
                                             f"{df.loc[random.randint(0, len(df)-1)]['gloss']} > {df.loc[random.randint(0, len(df)-1)]['spoken']}\n"
                                             f"{df.loc[random.randint(0, len(df)-1)]['gloss']} > {df.loc[random.randint(0, len(df)-1)]['spoken']}\n"},
                {"role": "user", "content": f"{line['gloss']}"},
                {"role": "assistant", "content": f"{line['spoken']}"}
            ]
        }
    )
    
```

```python
# print(row_data)

with open("0-shot.jsonl" , "w", encoding='utf-8') as f: 
	for i in row_data:  f.write(json.dumps(i, ensure_ascii=False) + "\n")
    
with open("0-shot.jsonl") as f: 
	for line in f: print(line)
```
