```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv('mydata.csv')
df = df.drop_duplicates()
df.rename(columns={'spoken language': 'spoken'}, inplace=True)
df.rename(columns={'gloss-level sentence': 'gloss'}, inplace=True)
df = df[['gloss', 'spoken']]
df
```

```python
replace_tuples = [
    ('.', ' .'), ('?', ' ?'), ('!', ' !'), ('  ', ' '), 
    ('얼마다', '얼마'), ('얼마이다', '얼마'), ('어디다', '어디'), 
    ('뭐이다', '무엇'), (' 뭐 ', ' 무엇 '), ('가능하다', '가능'), 
    ('하다 수 있다', '가능'), (' 수 있다', '가능'), ('적이다', '적'), 
    ('거이다', '것이다'), ('하다 하다', '하다')
]

for old, new in replace_tuples:
    df['gloss'] = df['gloss'].str.replace(old, new)

condition = ~df['gloss'].str.endswith(('.', '?', '!'))
df.loc[condition, 'gloss'] = df.loc[condition, 'gloss'] + " ."
df
```

```python
filtered_row = df[df['gloss'].str.contains('하다 안하다')]
filtered_row
```

```python
conditions = [
    df['gloss'].str.endswith('.'), 
    df['gloss'].str.endswith('?'),
    df['gloss'].str.endswith('!')
]
choices = [0, 1, 2]
df['label'] = np.select(conditions, choices, default=-1)
df
```

```python
df.to_csv('mydata(edited).csv', index=False)
```
