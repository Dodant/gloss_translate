```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv('mydata.csv')
df = df.drop_duplicates()
df.rename(columns={'spoken language': 'spoken'}, inplace=True)
df.rename(columns={'gloss-level sentence': 'gloss'}, inplace=True)
df = df[['gloss', 'spoken']]
df
```

```python
df['gloss'] = df['gloss'].str.replace('.', ' .')
df['gloss'] = df['gloss'].str.replace('?', ' ?')
df['gloss'] = df['gloss'].str.replace('!', ' !')
df['gloss'] = df['gloss'].str.replace('  ', ' ')
df['gloss'] = df['gloss'].str.replace('얼마다', '얼마')
df['gloss'] = df['gloss'].str.replace('얼마이다', '얼마')
df['gloss'] = df['gloss'].str.replace('어디다', '어디')
df['gloss'] = df['gloss'].str.replace('뭐이다', '무엇')
df['gloss'] = df['gloss'].str.replace(' 뭐 ', ' 무엇 ')
df['gloss'] = df['gloss'].str.replace('가능하다', '가능')
df['gloss'] = df['gloss'].str.replace('하다 수 있다', '가능')
df['gloss'] = df['gloss'].str.replace(' 수 있다', '가능')
df['gloss'] = df['gloss'].str.replace('적이다', '적')
df['gloss'] = df['gloss'].str.replace('거이다', '것이다')
df['gloss'] = df['gloss'].str.replace('하다 하다', '하다')
condition = ~df['gloss'].str.endswith(('.', '?', '!'))
df.loc[condition, 'gloss'] = df.loc[condition, 'gloss'] + " ."
df
```

```python
filtered_row = df[df['gloss'].str.contains('하다 안하다')]
filtered_row
```

```python
conditions = [
    df['gloss'].str.endswith('.'), 
    df['gloss'].str.endswith('?'),
    df['gloss'].str.endswith('!')
]
choices = [0, 1, 2]
df['label'] = np.select(conditions, choices, default=-1)
df
```

```python
df.to_csv('mydata_edited.csv', index=False)
```
